```python
import requests
from bs4 import BeautifulSoup
```

```python
# https://cgaxis.com/assets/?jsf=epro-archive-products&tax=assets:6038;product_cat:450
base_url = 'https://cgaxis.com/assets/?jsf=epro-archive-products&tax=assets:6038;'
main_url = base_url + 'product_cat:450&pagenum='
```

```python
# 제목 저장
title = []
```

```python
for i in range(1, 15):
  resp = requests.get(main_url, params={
      'pagenum': i
  })
  soup = BeautifulSoup(resp.text)
  product_lst = soup.select('h2.woocommerce-loop-product__title')

  for p in product_lst:
    title_lst = p.get_text()
    title.append(title_lst)
```

```python
print(len(title))
```

```python
import pandas as pd
```

```python
trees = []
df = pd.DataFrame(trees)
```

```python
df['title'] = title
```
